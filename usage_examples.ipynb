```julia
include("launcher.jl")
using .Launcher
using Plots
```

```julia
function plot_NE_spectral_function(data::NamedTuple)
    plot(
        data.results[1].E,
        data.results[1].Bω[:, 1, 1, 1, 1, 1, 1] .+ data.results[1].Bω[:, 2, 1, 1, 1, 1, 1],
        xlabel="energy [meV]",
        label="A(ε)↑ + A(ε)↓",
        title="Non-Equilibrium Spectral Function",
        legend=:topright,
        framestyle=:box,
        grid=false
    )
    plot!(
        data.results[1].E,
        data.results[1].Bω[:, 2, 1, 1, 1, 1, 1],
        label=" A(ε)↓",
    )
    plot!(
        data.results[1].E,
        data.results[1].Bω[:, 1, 1, 1, 1, 1, 1],
        label=" A(ε)↑",
    )
end

function plot_E_spectral_function(data::NamedTuple)
    plot(
        data.results[1].E,
        data.results[1].Aω[:, 1, 1, 1, 1, 1, 1] .+ data.results[1].Aω[:, 2, 1, 1, 1, 1, 1],
        xlabel="energy [meV]",
        label="A(ε)↑ + A(ε)↓",
        title="Equilibrium Spectral Function",
        legend=:topright,
        framestyle=:box,
        grid=false,
    )
    plot!(
        data.results[1].E,
        data.results[1].Aω[:, 2, 1, 1, 1, 1, 1],
        label=" A(ε)↓",
    )
    plot!(
        data.results[1].E,
        data.results[1].Aω[:, 1, 1, 1, 1, 1, 1],
        label=" A(ε)↑",
    )
end
```

# One Level

<img src="images/One_Level.svg" alt="Description" style="background:white; padding:4px;" />

## Model : CPU One level No U 

```julia
model = Launcher.init_model(device="cpu", level=1, U_interaction=false)
data, indexes = Launcher.init_parameters_simulation(
    model,
    parameters=(
        Tr=(3.0,),
        U=model.U_interaction ? (10.0,) : (0.0,),
        λ=(0.5,),
        eV=(-10.0,),
        ϵ=(0.0,)
    ),
    E=(start=-10.0, stop=10.0, length=500)
)
Launcher.run_simulation(
    model,
    data,
    indexes
)
indexes = nothing
```

```julia
plot(
    plot_NE_spectral_function(data), 
    plot_E_spectral_function(data),
    size=(1500, 500),
    margin= 10Plots.mm
)
```
